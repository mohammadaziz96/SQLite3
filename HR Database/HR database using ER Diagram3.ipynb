```python
import sqlite3
import pandas as pd
```

```python
connection=sqlite3.connect("hr")
```

```python
#df1=pd.read_sql("select * from employees limit 10", connection)
#df1
```

```python
#1 Write a query to list the number of jobs available in the employees table.
```

```python
df1=pd.read_sql("select count(distinct job_id) from employees", connection)
df1
```

```python
#2 Write a query to get the total salaries payable to employees.
```

```python
df2=pd.read_sql("select sum(salary) as 'Total Salary' from employees", connection)
df2
```

```python
#3 Write a query to get the minimum salary from employees table.
```

```python
df3=pd.read_sql("select min(salary) as 'Minimum Salary' from employees", connection)
df3
```

```python
#4 Write a query to get the maximum salary of an employee working as a Programmer.
```

```python
df4=pd.read_sql("select max(salary) as 'Maximum Salary' from employees", connection)
df4
```

```python
#5 Write a query to get the average salary and number of employees working the department 90.
```

```python
df5=pd.read_sql("select avg(salary) as 'Average Salary', count(employee_id) as 'No of Employee' from employees where department_id=90", connection)
df5
```

```python
#6 Write a query to get the highest, lowest, sum, and average salary of all employees.
```

```python
df6=pd.read_sql("select max(salary) as 'MAximum Salary', min(salary) as 'Minimum Salary', sum(salary) as 'Sum', round(avg(salary), 2) as 'Average Salary' from employees", connection)
df6
```

```python
#7 Write a query to get the number of employees with the same job.
```

```python
df7=pd.read_sql("select count(employee_id) as 'No of Employee', job_id as 'Job Assigned' from employees group by job_id", connection)
df7
```

```python
#8 Write a query to get the difference between the highest and lowest salaries. 
```

```python
df8=pd.read_sql("select max(salary)-min(salary) from employees", connection)
df8
```

```python
#9 Write a query to find the manager ID and the salary of the lowest-paid employee for that manager.
```

```python
df9=pd.read_sql("select first_name|| ' ' ||last_name as 'Employee Name', manager_id, min(salary) from employees group by manager_id", connection)
df9
```

```python
#10 Write a query to get the department ID and the total salary payable in each department.
```

```python
df10=pd.read_sql("select department_id, sum(salary) as 'Total Salary' from employees group by department_id", connection)
df10
```

```python
#11 Write a query to get the average salary for each job ID excluding programmer.
```

```python
df11=pd.read_sql("select avg(salary), job_id from employees where job_id != 'IT_PROG' group by job_id", connection)
df11
```

```python
#12 Write a query to get the total salary, maximum, minimum, average salary of employees (job ID wise), 
# for department ID 90 only.
```

```python
df12=pd.read_sql("select sum(salary), max(salary), min(salary), avg(salary), job_id, department_id from employees where department_id = 90 group by job_id", connection)
df12
```

```python
#13 Write a query to get the job ID and maximum salary of the employees where maximum salary 
# is greater than or equal to $4000.
```

```python
df13=pd.read_sql("select job_id, max(salary) from employees group by job_id having max(salary)>=4000", connection)
df13
```

```python
#14 Write a query to get the average salary for all departments employing more than 10 employees.
```

```python
df14=pd.read_sql("select job_id, avg(salary), count(*) from employees group by department_id having count(*)>10", connection)
df14
```
